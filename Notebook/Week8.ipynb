```python
# nl.py
# Prints each line of a file with a line number

import sys

# Check that a filename was provided
if len(sys.argv) < 2:
    print("Usage: python nl.py <filename>")
else:
    filename = sys.argv[1]

    # Open the file and read line by line
    with open(filename, "r") as file:
        line_number = 1
        for line in file:
            # Print the line number and the line contents
            print(f"{line_number}: {line.rstrip()}")
            line_number += 1
```

```python
# diff.py
# Compares two files and reports whether they are identical

import sys

# Check that two filenames were provided
if len(sys.argv) < 3:
    print("Usage: python diff.py <file1> <file2>")
else:
    file1 = sys.argv[1]
    file2 = sys.argv[2]

    # Read both files completely
    with open(file1, "r") as f1, open(file2, "r") as f2:
        contents1 = f1.read()
        contents2 = f2.read()

    # Compare the contents
    if contents1 == contents2:
        print("The files are the same.")
    else:
        print("The files are different.")
```

```python
# grep.py
# Prints lines from a file that contain a given search string

import sys

# Check that a search string and filename were provided
if len(sys.argv) < 3:
    print("Usage: python grep.py <search_string> <filename>")
else:
    search_string = sys.argv[1]
    filename = sys.argv[2]

    # Open the file and search line by line
    with open(filename, "r") as file:
        for line in file:
            if search_string in line:
                print(line.rstrip())
```

```python
# wc.py
# Counts the number of lines and characters in a file

import sys

# Check that a filename was provided
if len(sys.argv) < 2:
    print("Usage: python wc.py <filename>")
else:
    filename = sys.argv[1]

    with open(filename, "r") as file:
        lines = file.readlines()

    # Number of lines
    line_count = len(lines)

    # Number of characters (including newlines)
    char_count = sum(len(line) for line in lines)

    print("Lines:", line_count)
    print("Characters:", char_count)
```

```python
# spell.py
# Prints words in a file that are not found in a dictionary

import sys
import string

# Check that a filename was provided
if len(sys.argv) < 2:
    print("Usage: python spell.py <filename>")
    sys.exit()

filename = sys.argv[1]

# Load dictionary words into a set for fast lookup
# Adjust this path if needed (especially on Windows)
with open("words.txt", "r") as dict_file:
    dictionary = set(word.strip().lower() for word in dict_file)

# Open the text file to spell-check
with open(filename, "r") as file:
    text = file.read().lower()

# Remove punctuation
for char in string.punctuation:
    text = text.replace(char, "")

# Split text into words
words = text.split()

# Find and print words not in the dictionary
misspelled = set()

for word in words:
    if word not in dictionary:
        misspelled.add(word)

# Display results
print("Misspelled words:")
for word in sorted(misspelled):
    print(word)
```
