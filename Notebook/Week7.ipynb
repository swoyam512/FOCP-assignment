```python
# Function that returns a sorted list of unique letters in a string
def unique_letters(text):
    # Convert the string into a set to remove duplicates
    letters = set(text)

    # Convert the set back to a list and sort it
    return sorted(letters)


# Test the function
word = input("Enter a word: ")
print(unique_letters(word))
```

```python
# Letters that appear in at least one of the two words
def letters_in_either(word1, word2):
    return sorted(set(word1) | set(word2))


# Letters that appear in both words
def letters_in_both(word1, word2):
    return sorted(set(word1) & set(word2))


# Letters that appear in either word, but not in both
def letters_in_one_only(word1, word2):
    return sorted(set(word1) ^ set(word2))


# Test the functions
w1 = input("Enter first word: ")
w2 = input("Enter second word: ")

print("Either word:", letters_in_either(w1, w2))
print("Both words:", letters_in_both(w1, w2))
print("One only:", letters_in_one_only(w1, w2))
```

```python
# Dictionary to store countries and their capitals
capitals = {}

while True:
    # Ask the user to enter a country name
    country = input("Enter a country (or press Enter to quit): ")

    # End the program if Enter is pressed
    if country == "":
        break

    # Convert country name to lowercase for consistency
    country_key = country.lower()

    # Check if the capital is already known
    if country_key in capitals:
        print("The capital of", country, "is", capitals[country_key])
    else:
        # Ask the user for the capital if unknown
        capital = input("I don't know that. What is the capital? ")
        capitals[country_key] = capital
        print("Thank you, I've learned something new!")
```

```python
# Ask the user for a message
message = input("Enter a message: ")

# Convert message to lowercase so case does not matter
message = message.lower()

# Dictionary to store letter counts
counts = {}

# Initialise counts for letters a to z
for letter in "abcdefghijklmnopqrstuvwxyz":
    counts[letter] = 0

# Count occurrences of each letter in the message
for char in message:
    if char.isalpha():
        counts[char] += 1

# Convert dictionary items to a list of (letter, count) pairs
count_list = list(counts.items())

# Sort the list by count (highest first)
count_list.sort(key=lambda item: item[1], reverse=True)

# Display the six most common letters
print("Six most common letters:")
for i in range(6):
    letter, count = count_list[i]
    print(letter, ":", count)
```
