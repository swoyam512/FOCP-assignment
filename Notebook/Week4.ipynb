```python
# Function to check if a number is between 0 and 100 inclusive
def valid_number(num):
    return 0 <= num <= 100


# Test the function
number = int(input("Enter a number: "))

if valid_number(number):
    print("The number is valid.")
else:
    print("The number is not valid.")
```

```python
# Function that counts uppercase and lowercase letters in a string
def count_case(text):
    upper = 0
    lower = 0

    for char in text:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1

    return upper, lower


# Test the function
message = input("Enter some text: ")
uppercase, lowercase = count_case(message)

print("Uppercase letters:", uppercase)
print("Lowercase letters:", lowercase)
```

```python
# Ask the user for their name
name = input("Hello, what is your name? ")

# Capitalise the first letter and make the rest lowercase
formatted_name = name.capitalize()

# Display the greeting
print(f"Hello, {formatted_name}. Good to meet you!")
```

```python
# Function that removes the last character from a string
def remove_last_char(text):
    # If the string has one or zero characters, return it unchanged
    if len(text) <= 1:
        return text
    else:
        return text[:-1]


# Test the function
word = input("Enter some text: ")
print("Result:", remove_last_char(word))
```

```python
# Convert Celsius to Fahrenheit
def c_to_f(celsius):
    return (celsius * 9 / 5) + 32


# Convert Fahrenheit to Celsius
def f_to_c(fahrenheit):
    return (fahrenheit - 32) * 5 / 9


# Test the functions
c = float(input("Enter temperature in Celsius: "))
print("In Fahrenheit:", c_to_f(c))

f = float(input("Enter temperature in Fahrenheit: "))
print("In Celsius:", f_to_c(f))
```

```python
# Ask the user to enter a temperature ending with C
temp_input = input("Enter a temperature (e.g. 32C): ")

# Remove the last character and convert the number part to float
celsius = float(temp_input[:-1])

# Convert to Fahrenheit
fahrenheit = (celsius * 9 / 5) + 32

# Display the result
print(f"{fahrenheit}F")
```

```python
import statistics  # Used to calculate the mean

temperatures = []

# Read 6 temperature values
for i in range(6):
    temp_input = input("Enter a temperature (e.g. 20C): ")
    celsius = float(temp_input[:-1])
    temperatures.append(celsius)

# Display results
print("Maximum:", max(temperatures))
print("Minimum:", min(temperatures))
print("Mean:", statistics.mean(temperatures))
```

```python
# Keep asking for temperatures until the user presses Enter
while True:
    temp_input = input("Enter a temperature (or press Enter to finish): ")

    # Stop if the user presses Enter
    if temp_input == "":
        break

    # Convert input to a number (remove the C)
    celsius = float(temp_input[:-1])
    temperatures.append(celsius)

# Only calculate results if at least one value was entered
if len(temperatures) > 0:
    print("Maximum:", max(temperatures))
    print("Minimum:", min(temperatures))
    print("Mean:", statistics.mean(temperatures))
else:
    print("No temperatures entered.")
```
