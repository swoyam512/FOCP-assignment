```python
# Function that converts a positive integer to its binary representation
def to_binary(number):
    # bin() returns a string like '0b1010'
    # [2:] removes the '0b' prefix
    return bin(number)[2:]


# Test the function
num = int(input("Enter a positive integer: "))
print("Binary:", to_binary(num))
```

```python
# Function that returns all factors of an integer
def get_factors(n):
    factors = []

    # Check each number from 1 to n
    for i in range(1, n + 1):
        # If i divides n exactly, it is a factor
        if n % i == 0:
            factors.append(i)

    return factors


# Test the function
number = int(input("Enter an integer: "))
print("Factors:", get_factors(number))
```

```python
# Function that checks whether a number is prime
def is_prime(n):
    # Numbers less than or equal to 1 are not prime
    if n <= 1:
        return False

    # Check for any divisor other than 1 and itself
    for i in range(2, n):
        if n % i == 0:
            return False

    # If no divisors were found, the number is prime
    return True


# Test the function
number = int(input("Enter an integer: "))

if is_prime(number):
    print("The number is prime.")
else:
    print("The number is not prime.")
```

```python
# Function that removes spaces and reverses the message
def encrypt_simple(message):
    # Remove all spaces from the message
    message_no_spaces = message.replace(" ", "")

    # Reverse the string
    encrypted = message_no_spaces[::-1]

    return encrypted


# Test the function
text = input("Enter a message: ")
print("Encrypted message:", encrypt_simple(text))
```

```python
import random
import string

# Function that hides a message in random letters
def encrypt_hidden(message):
    # Remove spaces from the message
    message = message.replace(" ", "")

    # Generate a random interval between 2 and 20
    interval = random.randint(2, 20)

    encrypted = ""

    # Add each message character followed by random letters
    for char in message:
        encrypted += char

        # Fill the gap with random lowercase letters
        for i in range(interval - 1):
            encrypted += random.choice(string.ascii_lowercase)

    # Return the encrypted message and the interval used
    return encrypted, interval


# Test the function
msg = input("Enter a message to encrypt: ")
encrypted_msg, interval_used = encrypt_hidden(msg)

print("Encrypted message:", encrypted_msg)
print("Interval used:", interval_used)
```

```python
# Function that extracts the hidden message
def decrypt_hidden(encrypted_message, interval):
    decrypted = ""

    # Take every nth character based on the interval
    for i in range(0, len(encrypted_message), interval):
        decrypted += encrypted_message[i]

    return decrypted


# Test the function
encrypted = input("Enter the encrypted message: ")
interval = int(input("Enter the interval used: "))

print("Decrypted message:", decrypt_hidden(encrypted, interval))
```
