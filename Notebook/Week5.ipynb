```python
# Import the sys module
import sys

# sys.argv is a list of command-line arguments
# The first item is always the program name
argument_count = len(sys.argv) - 1

# Display the number of arguments
print("Number of arguments provided:", argument_count)
```

```python
# Import the sys module
import sys

# Get the list of arguments (excluding the program name)
arguments = sys.argv[1:]

# Check that at least one argument was provided
if len(arguments) == 0:
    print("No arguments provided.")
else:
    # Sort arguments by length
    arguments.sort(key=len)

    # Print the shortest argument
    print("Shortest argument:", arguments[0])
```

```python
# Import required modules
import sys
import urllib.request

# Check that a URL was provided
if len(sys.argv) < 2:
    print("Please provide a URL.")
else:
    url = sys.argv[1]

    try:
        # Attempt to open the URL
        response = urllib.request.urlopen(url)

        # Get the HTTP response code
        status_code = response.getcode()

        # Check if the response indicates success
        if status_code == 200:
            print("The website is working (HTTP 200).")
        else:
            print("Website responded with code:", status_code)

    except:
        # If an error occurs, the site is not reachable
        print("The website is not reachable.")
```

```python
# Import required modules
import sys
import statistics

# Get temperature values from the command line
arguments = sys.argv[1:]

# Check if any arguments were provided
if len(arguments) == 0:
    print("No temperature values provided.")
else:
    temperatures = []

    # Convert arguments to floats
    for value in arguments:
        temperatures.append(float(value))

    # Display results
    print("Maximum:", max(temperatures))
    print("Minimum:", min(temperatures))
    print("Mean:", statistics.mean(temperatures))
```

```python
# Import required modules
import sys
import shutil

# Check that a filename was provided
if len(sys.argv) < 2:
    print("Please provide a filename to back up.")
else:
    original_file = sys.argv[1]

    # Create a backup filename
    backup_file = original_file + ".bak"

    try:
        # Copy the file contents to the backup file
        shutil.copyfile(original_file, backup_file)
        print("Backup created:", backup_file)

    except FileNotFoundError:
        print("The specified file does not exist.")
```
