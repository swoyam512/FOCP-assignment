```python
# Ask the user to enter a password
password1 = input("Enter a new password: ")

# Ask the user to enter the password again
password2 = input("Re-enter the password: ")

# Check if both passwords match
if password1 == password2:
    print("Password set.")
else:
    print("Error: passwords do not match.")
```

```python
# Ask the user to enter a password
password1 = input("Enter a new password: ")
password2 = input("Re-enter the password: ")

# Check if passwords match
if password1 != password2:
    print("Error: passwords do not match.")

# Check if password length is valid
elif len(password1) < 8 or len(password1) > 12:
    print("Error: password must be between 8 and 12 characters.")

else:
    print("Password set.")
```

```python
# List of common passwords that are not allowed
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

# Ask for the password twice
password1 = input("Enter a new password: ")
password2 = input("Re-enter the password: ")

# Check if passwords match
if password1 != password2:
    print("Error: passwords do not match.")

# Check password length
elif len(password1) < 8 or len(password1) > 12:
    print("Error: password must be between 8 and 12 characters.")

# Check if password is in the bad password list
elif password1 in BAD_PASSWORDS:
    print("Error: password is too common.")

else:
    print("Password set.")
```

```python
# List of common passwords that are not allowed
BAD_PASSWORDS = ['password', 'letmein', 'sesame', 'hello', 'justinbieber']

# Loop until a valid password is set
while True:
    password1 = input("Enter a new password: ")
    password2 = input("Re-enter the password: ")

    # Check if passwords match
    if password1 != password2:
        print("Error: passwords do not match.")
    
    # Check password length
    elif len(password1) < 8 or len(password1) > 12:
        print("Error: password must be between 8 and 12 characters.")
    
    # Check against common passwords
    elif password1 in BAD_PASSWORDS:
        print("Error: password is too common.")
    
    # If all checks pass, exit the loop
    else:
        print("Password set successfully.")
        break
```

```python
# Ask the user for the table number
table = int(input("Enter a number between 0 and 12: "))

# Check if the number is valid
if 0 <= table <= 12:
    for i in range(13):
        print(f"{i} x {table} = {i * table}")
else:
    print("Error: number must be between 0 and 12.")
```

```python
# Ask the user for the table number
table = int(input("Enter a number: "))

# If the number is negative, print the table backwards
if table < 0:
    table = abs(table)  # Convert to positive
    for i in range(12, -1, -1):
        print(f"{i} x {table} = {i * table}")

# If the number is between 0 and 12, print normally
elif 0 <= table <= 12:
    for i in range(13):
        print(f"{i} x {table} = {i * table}")

else:
    print("Error: number must be between -12 and 12.")
```
